```python
from pprint import pprint as pp
patient_text = "A 21-year-old Indian female patient admitted to hospital  on 21-may-2010, who had increased bleeding from the scaling near her rectum , increased scaling , itching, increased lesion sites on back of her head, ears and legs. while receiving Enbrel. No historical medical condition was reported. The patient's current medical condition included psoriatic arthritis. No concomitant medications were provided. No co-suspect medications were reported. The patient began Enbrelon an unknown date, over ten years ago. On an unknown date, past year or two, the patient had increased bleeding from the scaling near his rectum after wiping herself, increased lesion sites on back of her head, ears and legs. On an unknown date in 2017, for a month, the patient stopped Enbrel therapy due to financial reasons and started back this past summer. On an unknown date in 2017, two and half months prior to this report, the patient had increased scaling and itching. No treatment information was received. The outcome of the events skin hemorrhage, skin exfoliation, pruritus, skin lesion was reported as unknown. Action taken with Enbrel was reported as unknown for the events skin exfoliation, pruritus, skin hemorrhage and skin lesion. The causal relationship between the events skin exfoliation, pruritus, skin hemorrhage, skin lesion and Enbrel was not provided by the consumer. Follow up has been requested."
pp(patient_text)
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
import os
current_dir = os.path.dirname("/NLP_Pharma_Complaints/word_cloud")

def create_cloud(text):
    stopwords = set(STOPWORDS)
    wc= WordCloud(background_color  = "white",
                 max_words=200,stopwords = stopwords)
    wc.generate(text)
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
    
    # Save word_cloud to a file
    wc.to_file(os.path.join(current_dir,"patientwc.png"))

create_cloud(patient_text)

    
```

```python
#What is gensim?
#Popular open-source NLP library
#Uses top academic models to perform complex tasks, Building document or word vectors
#Performing topic identification and document comparison
# Gensim creates  a Word- Vector which tells us the relation ship between the words
# Create a gensim dictionary and a corpus 
import gensim
import sklearn
from gensim.summarization import summarize
from gensim.summarization import keywords
print(summarize(patient_text, word_count=100))
print(" \nKeywords for this text are:","\n", keywords(patient_text))

# here we get the general summary that "There is no clear realtionship seen between the drug{Enbrel} and the sideeffect symptons"
# Hence one needs to be cautious before come to any conclusion just on the basis of the compalint text.
```

```python
# Knowing the part of speech is important if you want to find out the person or object the sentence is said about

# Noun_phrases - it helps us to know who or what the text is refering to
for np in blob.noun_phrases:
    print(np)
# One good thing we see in here is multiword extraction eg 'unknown date', 'financial reasons',' concomitant medication' etc
```

```python
# Next we will try to find verbs in the text that will help us
#  To know how or actions done
 # ['VB','VBZ','VBP','VBN','VBG','VBD'] 
    #VBZ	verb, 3rd person sing. present	takes
    #VBP	verb, sing. present, non-3d	take
    #VBD	verb, past tense	took
    #
for word,verb in blob.pos_tags:
    if verb in ['VB','VBZ','VBP','VBN','VBG','VBD']:
        print(f'{word} ==> {verb}')
```

```python
# Similarly we can generate a bog of words  that will give us a set of nouns
# ['NN','NNP']
#NN	noun, singular 'desk'
#NNS	noun plural	'desks'
#NNP	proper noun, singular	'Harrison'
#NNPS	proper noun, plural	'Americans'

for word,noun in blob.pos_tags:
    if noun in ['NN','NNP']:
        print(f'{word} ==> {noun}')
```
